```python
import numpy as np
from matplotlib import pyplot as plt
```

Prize-to-door assignments

```python
N=100000
m=7
p=3
#N trials, m doors for each trial, p doors opened by the host
sequence=np.ones((N,m),dtype=np.int_)
arr=np.arange(m)
for i in np.arange(N):
    np.random.shuffle(arr)
    sequence[i]=arr
```

Picks

```python
conservative=np.random.randint(3,size=N)#conservative player's draws
#host opens a door hiding a goat
host=np.ones(N,dtype=np.int_)
for i in np.arange(N):
    #random number drawn from those remaining upon ignoring the one alread picked by the player and the winning one. N.B.: the two may coincide
    a=np.random.randint(len(sequence[i][(sequence[i]!=conservative[i]) & (sequence[i]!=sequence[i][0])]))
    host[i]=sequence[i][(sequence[i]!=conservative[i]) & (sequence[i]!=sequence[i][0])][a]
#switcher player changes their bet
switcher=np.ones(N,dtype=np.int_)
for i in np.arange(N):
    #b=np.random.randint(1)
    b=np.random.randint(len(sequence[i][(sequence[i]!=conservative[i]) & (sequence[i]!=host[i])]))
    switcher[i]=sequence[i][(sequence[i]!=conservative[i]) & (sequence[i]!=host[i])][b]
#newcomer enters and places a bet
newcomer=np.ones(N,dtype=np.int_)
for i in np.arange(N):
    c=np.random.randint(len(sequence[i][sequence[i]!=host[i]]))
    newcomer[i]=sequence[i][sequence[i]!=host[i]][c]
```

Win ratios

```python
#record wins
#conservative case
conservative_wins_count=0
for i in np.arange(N):
    if sequence[i][0]==conservative[i]:
        conservative_wins_count += 1
conservative_wins_fraction=conservative_wins_count/N
#switcher case
switcher_wins_count=0
for i in np.arange(N):
    if sequence[i][0]==switcher[i]:
        switcher_wins_count += 1
switcher_wins_fraction=switcher_wins_count/N
#newcomer case
newcomer_wins_counts=0
for i in np.arange(N):
    if sequence[i][0]==newcomer[i]:
        newcomer_wins_counts += 1
newcomer_wins_fraction=newcomer_wins_counts/N
print("Conservative player's fraction is {}, switcher player's fraction is {}, newcomer's fraction is {}".format(conservative_wins_fraction,switcher_wins_fraction,newcomer_wins_fraction))
```

```python
conservative=np.random.randint(m,size=N)#conservative player's draws
#host opens p doors hiding a goat
host=np.ones((N,p),dtype=np.int_)
for i in np.arange(N):
    #p random numbers drawn from those remaining upon ignoring the one alread picked by the player and the winning one. N.B.: the two may coincide
    a=np.random.default_rng().choice(len(sequence[i][(sequence[i]!=conservative[i]) & (sequence[i]!=sequence[i][0])]),size=p,replace=False)
    host[i]=sequence[i][(sequence[i]!=conservative[i]) & (sequence[i]!=sequence[i][0])][a]
#switcher player changes their bet
switcher=np.ones(N,dtype=np.int_)
for i in np.arange(N):
    #b=np.random.randint(1)
    b=np.random.randint(len(sequence[i][(sequence[i]!=conservative[i]) & (sequence[i]!=host[i])]))
    switcher[i]=sequence[i][(sequence[i]!=conservative[i]) & (sequence[i]!=host[i])][b]
#newcomer enters and places a bet
newcomer=np.ones(N,dtype=np.int_)
for i in np.arange(N):
    c=np.random.randint(len(np.setdiff1d(sequence[4],host[4])))
    newcomer[i]=sequence[i][sequence[i]!=host[i]][c]
```
